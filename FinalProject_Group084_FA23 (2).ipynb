# Overview / Abstract

The purpose of this project is to examine whether there is a correlation between the sales of shooter video games and the occurrence of gun-related crimes in the United States from 2014 to 2016. Additionally, this project inquires whether there is a statistically significant difference in the average number of crimes that happen in the 30 days following the release of popular shooter video games compared to the average number of crimes in any 30-day period overall. Through an OLS analysis and a Mann-Whitney U test, conducted using datasets on popular video games, gun violence incidents in the United States, and video game sales, our findings support the hypothesis that there is no correlation between these factors and no statistically significant difference in the number of crimes following the release of popular shooter video games. 

# Research Question

From 2014 to 2016 in the United States, is there a relationship between how many sales (in millions) shooter video games make in a year and how many gun-related crimes occur in the U.S. in that same year? Furthermore, is there a statistical difference between the average number of crimes that occur in the 30-day periods following the release of popular large shooter game (defined by over 2 million U.S. sales) and the average number of crimes that occur in any 30-day period overall? 

## Background and Prior Work

Video games with violent subject matters, such as those that include firearms, gore, and combat, have been controversial in the public eye for decades now. In 1993, governments met to ban the incendiary Sega game “Night Trap”<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1); the conversation has only escalated since then, with continuing discussion on how and if video games might inspire mass shootings in the 21st century<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). 

The field of psychology is particularly interested in the topic. Although some studies do confirm that violent video games can cause more aggression in players than non-violent video games<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3), other studies provide deeper insight by suggesting that this increased aggressive behavior only lasts between 4 to 9 minutes<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). This offers some insight on societal implications and possible effect sizes of violent video games on aggression.

A gap in existing research, however, is systematically observing the relationship between video games and violence, specifically crimes involving gun violence, in the real world. Due to ethical and practical concerns, this is not something that can be empirically tested in a lab setting. As gun violence is growing as a concern in the United States with 2021 holding the record for the most gun-related injuries in the country<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5), we believe it would be valuable to investigate if a connection between popular shooter games being released (and therefore played), and gun-related crimes exists by examining available data.

# Hypothesis


Our hypothesis is that the release of shooter video games has no correlation with the occurrence of violent gun-related crimes across the country. We believe that a variety of factors influence violent gun-related crimes, such as mental health issues, gang activity, and so on, with shooter video games being one of the least influential. Furthermore, even if there is statistically significant aggression that arises from playing shooter games — as stated in previous psychological studies, we believe the *effect size* of that aggression to be too small to lead to these crimes.

Ok, let's address a different perspective with these two factors (shooter video games and gun crime incidents). What does the incident frequency in the days following a popular shooter video game release look like?

```python
# Taking the popular video games, taking the amount of crimes within 30 days of its release, and adding it to a new column.
pop_vg['gun_crimes_count'] = 0

pop_vg['date'] = pd.to_datetime(pop_vg['date'])

for index, game in pop_vg.iterrows():
    release_date = game['date']
    
    #create a variable that calculates the range of 30 days
    timedelta_array = pd.to_timedelta(np.arange(30), 'D')
    stdev_date = (release_date + timedelta_array).astype(str)
    
    # Filter gun crimes within 30 days of the game release
    mask = (crimes['date'] >= release_date) & (crimes['date'] <= stdev_date.max())
        
    # Update the count in the new column
    pop_vg.at[index, 'gun_crimes_count'] = len(crimes[mask]) 
    
print(pop_vg)
```

```python
#Sort pop_vg by date 

pop_vg = pop_vg.sort_values(by = 'date')
pop_vg
```

```python
#Create a bar plot

plt.bar(pop_vg['Name'], pop_vg['gun_crimes_count'], width = 0.5)
plt.title('Gun Crime Counts Within 30 Days of Video Game Releases')
plt.xlabel('Release Date')
plt.ylabel('Gun Crime Count')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

The barplot above illustrates the number of gun crimes that occurred within 30 days after each game was released. The graph doesn't seem to have an observable pattern in the distribution of gun crime counts following game releases, and it seems that these popular games all had approximately equal frequency of crimes that occur after their release. The relationship between video game launches and gun crimes is not clear based on this plot alone.

```python
display(pop_vg['gun_crimes_count'].describe())

# What's the average for the 30 day period?
vg_m_avg = pop_vg['gun_crimes_count'].describe()['mean']
print()
print('Here\'s our average crime count for the 30 day period following a popular shooter video game release')
vg_m_avg
```

Seems like the crimes that occur following a game release averages around 4368. What's the average for any 30 day period?

```python
# We can find how many crimes happen during a 30 day period in general.

# We can start by using our crimes_d dataframe, which has the count of crimes every day.
# Let's generate 30 random observations (to represent a random 30 day period), and sum the incidents that happen across that period.
# Let's do this 6 times to make a dataframe with the same number of observations as 'pop_vg'.

crimes_d_30 = pd.DataFrame(columns = ['count'])

i = 0
while i < 6:
    temp_df = {
        'count': [crimes_d['count'].sample(n=30).sum()]
    }
    temp_df = pd.DataFrame.from_dict(temp_df)
    crimes_d_30 = pd.concat([crimes_d_30, temp_df])
    i = i + 1

crimes_d_30 = crimes_d_30.reset_index(drop = True)
display(crimes_d_30)

crimes_d_avg = crimes_d_30['count'].mean()

print('the average incidents across 6 random 30-day period is ')
print(crimes_d_avg)
```

```python
# Let's plot the average of 'pop_vg' incidents and 'crimes_d_30' incidents. 
# Again, the former represents the average number of incidents in the 30 days following a video game release across 6 video games.
# The latter represents the average number of incidents in a random 30 days.
# Let's plot the two averages.

incidents_30_avg = {
    '30 day periods after vg release':[vg_m_avg],
    '30 day periods in general':[crimes_d_avg]
    
}
incidents_30_avg = pd.DataFrame.from_dict(incidents_30_avg)
incidents_30_avg

sns.barplot(
    data = incidents_30_avg
)

plt.title('Gun Crime Counts Over 30 Day Periods')
plt.xlabel('Period Type')
plt.ylabel('Gun Crime Count')
```

There're no *seemingly* significant difference between the two groups.

We will investigate *actual* significance (inferentially) in our next section.

Note: 'vg' stands for 'video games'.

### Inferential Tests

Let's revisit our research question and address them individually. 

1. From 2014 to 2016 in the United States, is there a relationship between how many sales (in millions) shooter video games make in a year and how many gun-related crimes occur in the U.S. in that same year?

#### Question 1: Relationship in Sales

We will run an OLS analysis for this section.

```python
import statsmodels.formula.api as stats

# Does the sale of video games predict gun violence incidents?
OLS_result = stats.ols(formula="incidents ~ NA_Sales", data=combined).fit()
# Remember this is what our 'combined' dataframe looks like - it looks across sales and incidents across our years of interest.
display(combined)

print(OLS_result.summary())
```

Seeing that the *p* value is very large (0.593), there is no statistical significant relationship between sales and crime incidents. Furthermore, R-squared is also quite low (0.355). This provides insight for our first research question.

#### Question 2: Differences in Crime Rates

We will run a Mann Whitney U test for this section, since our distributions are *not* normal for `pop_vg` and `crimes_d_30` incident counts.

```python
import scipy.stats as scipy

# Is the crime frequency after a video game release statistically different than normal crime frequency?

incidents_30 = pd.concat([pop_vg['gun_crimes_count'], crimes_d_30['count']], axis = 1)
incidents_30.reset_index(inplace = True, drop = True)
incidents_30.rename(
    columns = {'gun_crimes_count':'incidents 30 days within vg',
               'count':'incidents within random 30 days'},
    inplace = True
)
incidents_30 = incidents_30.astype(int)
display(incidents_30)

u, p = scipy.mannwhitneyu(
   x = incidents_30['incidents 30 days within vg'],
   y = incidents_30['incidents within random 30 days'],
    #axis = 1
)
print("our p value for the mann whitney u test is:")
print(p)
```

Seeing that the p value is very large (larger than 0.05), there is no statistical significant difference between the number of incidents following a game and the number of incidents normally. This provides insight for our second research question.
